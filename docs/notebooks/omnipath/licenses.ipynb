# Licenses in OmniPath databases

Most prior knowledge databases make their information accessible to all entities,
including academic, commercial, and nonprofit organizations.
However, some databases impose usage restrictions.

To simplify this, OmniPath allows filtering of results based on the selected license.
By default, the academic license is used, as it grants access to all available entries,
though this can be modified using the license parameter.

```python
import decoupler as dc

dc.op.collectri(license='academic')
```

Changing the license to non-profit or commercial may result in the
loss of certain entries.

```python
dc.op.collectri(license='nonprofit')
```

```python
dc.op.collectri(license='commercial')
```

For some resources, entries may not be available at all.

```python
dc.op.resource('HPMR', license='commercial')
```

The removal of entries is required to comply with licensing regulations,
and users are expected to be aware of and adhere to these guidelines.
